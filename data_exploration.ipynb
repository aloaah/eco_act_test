```python
import pandas as pd
```

```python
df = pd.read_excel("data/donnees_candidats_dev_python.xlsx")
```

```python
df.head()
```

```python
df.count()
```

```python
print(df.nunique())
```

```python
df = df.dropna(axis='columns', how='all') #drop empty columns
```

```python
df.columns = [col.replace(" ", "_").replace("'", "").replace("é", "e").lower() for col in df.columns]
```

```python
df.columns
```

```python
df['code_gaz_supplementaire_1'].fillna('', inplace=True)
df['valeur_gaz_supplementaire_1'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
additional_gases = df.pivot(columns='code_gaz_supplementaire_1', values='valeur_gaz_supplementaire_1')
```

```python
additional_gases.drop([""], axis=1, inplace=True)
```

```python
additional_gases
```

```python
additional_gases.columns = [f'gaz_{col}' for col in additional_gases.columns]
```

```python
df = df.join(additional_gases)
```

```python
df.drop(columns=['code_gaz_supplementaire_1', 'valeur_gaz_supplementaire_1'], inplace=True)
df
```

```python
df.nunique()
```

```python
df["code_de_la_categorie"] = df["code_de_la_categorie"].replace(" > ", "/", regex=True)
```
